```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

# Correlation and Simple Linear Regression

```python
d = {'Height': [60, 61, 61, 62, 62, 63, 63, 63, 64, 64, 65, 65, 65, 66, 66, 67, 67, 68, 68, 69], 'Weight': [92, 105, 100, 107, 114, 112, 120, 120, 122, 127, 122, 136, 129, 136, 132, 140, 146, 129, 134, 153]}
df = pd.DataFrame(data=d)
print(df)
```

```python
df.plot(x="Height", y="Weight", kind="scatter")
plt.show()
```

```python
H=df['Height'].values.reshape(-1, 1)
W=df['Weight'].values
model = LinearRegression().fit(H,W)
df.plot(x="Height", y="Weight", color='red', kind="scatter")
plt.plot(H, model.predict(H), color='blue', linewidth=3)
plt.show()
```

## The Simple Linear Regression Model

$$
y = \beta_0 + \beta_1 x
$$

where

$y$ is the dependent variable,

$x$ is the independent variable,

$\beta_0$ is the intercept, and

$\beta_1$ is the slope of the line

## Example 1.

Suppose we have the following linear regression model between the number years of education, denoted by $educ$, and the wage received per day in PHP, denoted by $wage$,  $$ wage = 610 + 350 educ.$$

* Here, $\beta_0 = 610$ and $\beta_1 = 350$.  
* The average daily wage for an individual with 0 years of education is $\beta_0$, which is 610 PHP.
* For every additional year of education, an individual's average daily wage is increased by 350 PHP, which is $\beta_1$.
* If I have 12 years of education, what will be my daily wage?


```python
educ = float(input("Your number of years of education: "))
from IPython.display import display, Math
display(Math(r'\text{This means, } educ=%.2f' % educ))
display(Math(r'\text{Substituting into the equation gives us, } wage= 610 + 350*%.2f' % educ))
wage = 610 + 350*educ
display(Math(r'\text{Therefore, your daily wage is %.2f PHP}' % wage))
```
